```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'C:\\Users\\Parag\\Downloads\\Electric_Vehicle_Population_Data (1).csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Basic Summary Statistics
print("Basic Summary Statistics:")
print(data.describe())

# Count of electric vehicles by Make and Model
print("\nCount of Vehicles by Make:")
print(data['Make'].value_counts())

print("\nCount of Vehicles by Model:")
print(data['Model'].value_counts())
```

```python
# Histogram of Model Year
plt.figure(figsize=(10, 6))
sns.histplot(data['Model Year'], kde=False, bins=30)
plt.title('Distribution of Model Years')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.show()

# Bar Chart of Electric Vehicle Types
plt.figure(figsize=(10, 6))
sns.countplot(x='Electric Vehicle Type', data=data)
plt.title('Count of Electric Vehicle Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'data' is your DataFrame
# Replace 'path_to_your_dataset.csv' with the actual path to your dataset
data = pd.read_csv('C:\\Users\\Parag\\Downloads\\Electric_Vehicle_Population_Data (1).csv')

# Filtering out non-numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=[np.number])

# 1. Boxplot for Electric Range by Vehicle Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Electric Vehicle Type', y='Electric Range', data=data)
plt.title('Electric Range by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Electric Range')
plt.show()

# 2. Count of Vehicles by County
plt.figure(figsize=(10, 6))
data['County'].value_counts().plot(kind='bar')
plt.title('Count of Vehicles by County')
plt.xlabel('County')
plt.ylabel('Count')
plt.show()

# 3. Electric Range vs. Model Year
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Model Year', y='Electric Range', data=data)
plt.title('Electric Range vs Model Year')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.show()

# 4. Base MSRP Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Base MSRP'], kde=True)
plt.title('Distribution of Base MSRP')
plt.xlabel('Base MSRP')
plt.ylabel('Frequency')
plt.show()

# 5. Heatmap of Correlations
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
